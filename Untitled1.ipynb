```python
from datasets import load_dataset
from huggingface_hub import snapshot_download
from PIL import Image
import os

# 1. Download full dataset snapshot
dataset_dir = snapshot_download(
    repo_id="Efficient-Large-Model/vila-q-data-train",
    repo_type="dataset"
)

print("Dataset snapshot at:", dataset_dir)

# 2. Load dataset (metadata only)
ds = load_dataset("Efficient-Large-Model/vila-q-data-train")

sample = ds["train"][0]

# 3. Save images locally
os.makedirs("examples/resonator", exist_ok=True)

for idx, rel_path in enumerate(sample["image"]):
    full_path = os.path.join(dataset_dir, rel_path)

    img = Image.open(full_path)
    img.save(f"examples/resonator/resonator_{idx}.png")

print("Saved files:", os.listdir("examples/resonator"))
```

```python
from datasets import load_dataset
from PIL import Image
import os

ds = load_dataset(
    "Efficient-Large-Model/vila-q-data-train",
    streaming=True
)

sample = next(iter(ds["train"]))

os.makedirs("examples/resonator", exist_ok=True)

for idx, img_path in enumerate(sample["image"]):
    # HF streaming gives file-like objects
    img = Image.open(img_path)
    img.save(f"examples/resonator/resonator_{idx}.png")

print("Saved files:", os.listdir("examples/resonator"))
```
